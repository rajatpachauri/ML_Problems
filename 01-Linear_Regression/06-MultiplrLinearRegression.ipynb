```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import sklearn.metrics as sm
```

```python
path = 'data/tv_show_viewers_datasest.csv'
dataset = pd.read_csv(path)
dataset.head()
```

```python
dataset.describe()
```

```python
def train_test():
    features = []
    target = []
    
    for char1, char2, char3, char4, char5, fight, comedy, romance, viewers in zip(dataset['Character1_appeared'],
                                                                                  dataset['Character2_appeared'],
                                                                                  dataset['Character3_appeared'],
                                                                                  dataset['Character4_appeared'],
                                                                                  dataset['Character5_appeared'],
                                                                                  dataset['Fight_scenes'],
                                                                                  dataset['Comedy_scences'],
                                                                                  dataset['Romance_scence'],
                                                                                  dataset['Viewers']):
        features.append([char1, char2, char3, char4, char5, fight, comedy, romance])
        target.append(viewers)
        
    
    return features, target
```

```python
x_train, y_train = train_test()
```
